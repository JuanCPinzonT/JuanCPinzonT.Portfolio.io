# How are in-demand skills trending for Data Analysts?

## Methology

1. Aggregate skill counts monthly
2. Re-analyze based on percentage of total jobs
3. Plot the monthly skill demand

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt
from debugpy.server.cli import in_range

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
## Skill count per month for data analysts
df_DA_ES = df[(df['job_title'] == 'Data Analyst') & (df['job_country'] == 'Spain')].copy()

df_DA_ES['job_posted_month_no'] = df_DA_ES['job_posted_date'].dt.month

# Explode job_skills column
df_DA_ES_explode = df_DA_ES.explode('job_skills')
```

```python
#Pivot the DF

df_DA_ES_pivot = df_DA_ES_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

# Create new row of totals to be able to sort the df
df_DA_ES_pivot.loc['total'] = df_DA_ES_pivot.sum()

df_DA_ES_pivot = df_DA_ES_pivot[df_DA_ES_pivot.loc['total'].sort_values(ascending=False).index]

df_DA_ES_pivot = df_DA_ES_pivot.drop('total')
```

```python
# Percent of the skill

df_DA_totals = df_DA_ES.groupby(['job_posted_month_no']).size()
```

```python
df_DA_ES_percent = df_DA_ES_pivot.div(df_DA_totals/100, axis=0)
```

```python
df_DA_ES_percent = df_DA_ES_percent.reset_index()
df_DA_ES_percent['job_posted_month'] = df_DA_ES_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_ES_percent = df_DA_ES_percent.set_index('job_posted_month')
df_DA_ES_percent = df_DA_ES_percent.drop(columns='job_posted_month_no')
df_DA_ES_percent
```

```python
# Plot the df for only the first 5 skills

df_plot = df_DA_ES_percent.iloc[:, :5]

sns.lineplot(data=df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending 5 top skills for Data Analysts in Spain')
plt.ylabel('Likelihood in job posing')
plt.xlabel('2023')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5, frameon=False)

from matplotlib.ticker import PercentFormatter
ax = plt.gca()
ax.yaxis.set_major_formatter(PercentFormatter(decimals=0))

# For practice only
for i in range(5):
    plt.text(11, df_plot.iloc[-1, i], df_plot.columns[i], fontsize=9)
```
